```python
Tempdf = df[['DATE','TAVG','TMAX', 'TMIN']]
```

```python
Tempdf.columns
```

```python
Tempdf.shape
```

```python
# count number of nulls in each column
Tempdf.isnull().sum()
```

```python
#filtering out the rows with null values in TMAX
Tempdf = Tempdf[Tempdf.TMAX.notnull()]
```

```python
# count number of nulls in each column
Tempdf.isnull().sum()
```

```python
#filtering out the rows with null values in TMIN
Tempdf = Tempdf[Tempdf.TMIN.notnull()]
```

```python
# count number of nulls in each column
Tempdf.isnull().sum()
```

```python
Tempdf.TAVG.isna().sum()
```

```python
Tempdf['d_TAVG'] = (Tempdf.TMIN + Tempdf.TMAX)/2
```

```python
Tempdf.head()
```

```python
# replacing missing values in TAVG with average of TMAX and TMIN
Tempdf['TAVG'] = Tempdf['TAVG'].fillna(Tempdf['d_TAVG'])
```

```python
Tempdf.head()
```

```python
Tempdf.shape
```

```python
Tempdf.groupby('DATE').count()
```

For each calendar day there are multiple rows (from different latitudes and longitudes). We will summarize temperature values and group them at the date level, to merge with the bike data set

# group the data frame on dates and take average of TAVG, TMAX, TMIN columns
AggTempdf = Tempdf.groupby('DATE').agg(
                                        'TAVG' : {'m_TAVG': 'mean'}, # average of TAVG for the day
                                        'TMAX' : {'m_TMAX': 'mean'}, # average of TMAX for the day
                                        'TMIN' : {'m_TMIN': 'mean'}, # average of TMIN for the day
                                        )

```python
# group the data frame on dates and take average of TAVG, TMAX, TMIN columns
AggTempdf = Tempdf.groupby('DATE', as_index=False).agg(
                                       { 
                                        'TAVG':  'mean', # average of TAVG for the day
                                        'TMAX' : 'mean', # average of TMAX for the day
                                        'TMIN' :  'mean', # average of TMIN for the day
                                       }
)
```

```python
AggTempdf.shape
```

```python
AggTempdf.groupby('DATE').count()
```

```python
AggTempdf.head()
```

Since we are considering bike data set from 2018 Jan, the data prior to that can be filtered out

```python
LATempdf = AggTempdf[AggTempdf['DATE'] > '2017-12-31']
```

```python
LATempdf.shape
```
